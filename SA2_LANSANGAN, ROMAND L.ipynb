II. MovieLens 25M Dataset (See Files for the MovieLens 25M Dataset folder)

```python
dfMovies = pd.read_csv('Movielens 25M Dataset/movies.csv')
```

1) How many movies are released each year? What is the percentage of change over time?

```python
dfMovies['year'] = dfMovies['title'].str.extract(r'\((\d{4})\)')
moviesGroupedByYearCount = dfMovies.groupby('year').count()
moviesGroupedByYearDrop = moviesGroupedByYearCount.drop(columns=['title', 'movieId'])
moviesGroupedByYearCount = moviesGroupedByYearDrop.rename(columns={'genres': 'count'})
moviesGroupedByYearCount.reset_index(inplace=True)
moviesGroupedByYearPctChange = moviesGroupedByYearCount['count'].pct_change().fillna(0) * 100
moviesGroupedByYearPctChange = moviesGroupedByYearPctChange.to_frame().rename(columns={'count': 'pct_change'})
moviesGroupedByYearPctChange['year'] = moviesGroupedByYearCount['year']
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(16, 6))
sns.barplot(data=moviesGroupedByYearCount, x='year', y='count', ax=ax1)
ax1.set_title('Movie Count by Year')
ax1.set_xlabel('Year (1874-2019)')
ax1.set_ylabel('Count')
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45)
sns.barplot(data=moviesGroupedByYearPctChange, x='year', y='pct_change', ax=ax2)
ax2.set_title('Percentage Change in Movie Count by Year')
ax2.set_xlabel('Year (1874-2019)')
ax2.set_ylabel('Percentage Change')
ax2.set_xticklabels(ax2.get_xticklabels(), rotation=45)
plt.tight_layout()
plt.show()
```

2) What are the most popular genre of movies?


```python
genreColumn = dfMovies['genres'].str.cat(sep='|').split('|')
genres = list(set(genreColumn))
genreCounts = {}
for genre in genres:
    count = dfMovies['genres'].str.contains(genre).sum()
    genreCounts[genre] = count

genreCounts_df = pd.DataFrame(list(genreCounts.items()), columns=['Genre', 'Count'])
sns.set(rc={'figure.figsize':(12, 8)})
sns.set_style('whitegrid')
sns.set_context('talk')
palette = sns.color_palette("husl", len(genres))
sns.barplot(data=genreCounts_df, x='Genre', y='Count', palette=palette)
plt.title('Movie Count by Genre')
plt.xlabel('Genre')
plt.ylabel('Movie Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

3) Does the taste of movies changes over time?


```python
dfMovies = pd.read_csv('Movielens 25M Dataset/movies.csv')
dfMovies['year'] = dfMovies['title'].str.extract(r'\((\d{4})\)')
genres = dfMovies['genres'].str.cat(sep='|').split('|')
genresList = list(set(genres))
genreCounts = {}
for genre in genresList:
    moviesGenre = dfMovies[dfMovies['genres'].str.contains(genre, regex=False)]
    yearlyCounts = moviesGenre.groupby('year')['genres'].count()
    genreCounts[genre] = yearlyCounts.to_dict()

allYears = set()
for genre in genreCounts:
    allYears.update(genreCounts[genre].keys())
sortedYears = sorted(allYears)

data = []
for genre in genreCounts:
    for year in sortedYears:
        if year in genreCounts[genre]:
            count = genreCounts[genre][year]
        else:
            count = 0
        data.append({'Year': year, 'Count': count, 'Genre': genre})
df_long = pd.DataFrame(data)
sns.set(rc={'figure.figsize':(12, 8)})
sns.set_style('whitegrid')
sns.set_context('talk')
palette = sns.color_palette("husl", len(genresList))
sns.lineplot(data=df_long, x='Year', y='Count', hue='Genre', style='Genre', markers=True, palette=palette)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title('Movie Count by Genre and Year')
plt.xlabel('Year (1874-2019)')
plt.ylabel('Movie Count')


plt.tight_layout()
plt.show()
```

4) What are the top 20 movies that has the top amount of users rating and has the highest rate?

```python
dfRatings = pd.read_csv('Movielens 25M Dataset/ratings.csv')
dfRatingsPerMovie = dfRatings.groupby('movieId')
dfRatingsTotalReviewers = dfRatingsPerMovie['rating'].count()
dfTotalRatingsRanking = dfRatingsTotalReviewers.sort_values(ascending=False)
top20Movies = dfTotalRatingsRanking.head(20).index
dfRatingsTotalReviewersFiltered = dfRatingsTotalReviewers.loc[top20Movies]
dfAvgRating = dfRatingsPerMovie['rating'].mean().loc[top20Movies].to_frame().rename(columns={'rating': 'average_rating'})
dfAvgRating = dfAvgRating.sort_values(by='average_rating', ascending=False)
dfAvgRating['ranking_rating'] = dfAvgRating['average_rating'].rank(ascending=False)
dfAvgRating = pd.merge(dfAvgRating, dfMovies[['movieId', 'title']], on='movieId', how='left')
dfAvgRating = pd.merge(dfAvgRating, dfRatingsTotalReviewersFiltered.to_frame().reset_index().rename(columns={'rating': 'total_reviewers'}), on='movieId', how='outer')
dfAvgRating['ranking_reviewers'] = dfAvgRating['total_reviewers'].rank(ascending=False, method='min')
dfAvgRating['total_reviewers'] = dfAvgRating['total_reviewers']
dfAvgRating[['title', 'ranking_rating', 'ranking_reviewers', 'total_reviewers', 'average_rating']].head(20)
```

```python
plt.figure(figsize=(12, 8))
reviewers_plot = sns.barplot(x='total_reviewers', y='title', data=dfAvgRating, palette='viridis', errorbar=None)
plt.xlabel('Total Reviewers', fontsize=14)
plt.ylabel('Movie Title', fontsize=14)
plt.title('Top 20 Movies by Total Reviewers: Number of Reviewers', fontsize=16)
plt.show()
plt.figure(figsize=(12, 8))
rating_plot = sns.barplot(x='average_rating', y='title', data=dfAvgRating, palette='viridis', errorbar=None)
plt.xlabel('Average Rating', fontsize=14)
plt.ylabel('Movie Title', fontsize=14)
plt.title('Top 20 Movies by Total Reviewers: Average Ratings', fontsize=16)
plt.show()
```

I did this for the sake of it. I filtered out all the movies outside of the 99th percentile then rank them based on the average rating. I thought this is the more apppropriate approach when finding the top 20 movies of all-time.

```python
dfRatingsPerMovie = dfRatings.groupby('movieId')
dfRatingsTotalRatings = dfRatingsPerMovie['rating'].sum()
dfRatingsTotalReviewers = dfRatingsPerMovie['rating'].count()
dfTotalRatingsRanking = dfRatingsTotalReviewers.sort_values(ascending=False)
dfTotalRatingsRanking99thPercentile = dfTotalRatingsRanking.quantile(0.99)
dfTotalRatingsRankingFiltered = dfTotalRatingsRanking[dfTotalRatingsRanking >= dfTotalRatingsRanking99thPercentile]
dfRatingsTotalReviewersFiltered = dfRatingsTotalReviewers[dfTotalRatingsRanking >= dfTotalRatingsRanking99thPercentile]
dfAvgRating = dfRatingsPerMovie['rating'].mean()[dfTotalRatingsRankingFiltered.index]
dfAvgRating = dfAvgRating.to_frame().rename(columns={'rating': 'average_rating'})
dfAvgRating = dfAvgRating.sort_values(by='average_rating', ascending=False)
dfAvgRating['ranking'] = dfAvgRating['average_rating'].rank(ascending=False)
dfAvgRating[['ranking', 'average_rating']]
dfAvgRating = pd.merge(dfAvgRating, dfMovies[['movieId', 'title']], on='movieId', how='left')
dfAvgRating = pd.merge(dfAvgRating, dfRatingsTotalReviewersFiltered.to_frame().reset_index().rename(columns={'rating': 'total_reviewers'}), on='movieId', how='outer')
dfAvgRating['total_reviewers'] = dfAvgRating['total_reviewers'].fillna(0)
dfAvgRating[['title', 'ranking', 'total_reviewers', 'average_rating']].head(20)
```

5) What is the most of popular season for releasing movies?

```python
dfRatings['month'] = pd.to_datetime(dfRatings['timestamp'], unit='s').dt.month
dfMonth = dfRatings.groupby('month').count()['rating']
dfMonth = dfMonth / 100000
plt.figure(figsize=(12, 8))
sns.barplot(x='month', y='rating', data=dfMonth.reset_index(), palette='viridis')
plt.xlabel('Month', fontsize=14)
plt.ylabel('Total Number of Reviews (in hundreds of thousands)', fontsize=14)
plt.title('Total Number of Reviews by Month', fontsize=16)
plt.ticklabel_format(style='plain', axis='y', scilimits=(6, 6))
plt.show()
```

I used the timestamp column in ratings.csv for there's no release dates of the movies directly in any dataset. I thought this is the most appropriate approach for it gives a good grasp when are much more people reviewing movies and therefore watching them.

As it turns out, according to the data presented above, the most popular season to release movies was from october (10) to december (12). Which is just about right for they indeed were holiday seasons. Though, I'm surprised, the most wasn't december, but we'll take what we can get.
