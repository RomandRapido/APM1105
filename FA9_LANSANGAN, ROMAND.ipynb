5) Fill NA in likes with 0, NA in comments with 0, and drop others.

7) Explore album type and display counts for album, single, and compilation.

```python
albumCounts = df['Album_type'].value_counts()
albumCounts
```

8) Create a pie chart for Album Types as a Title and make some remarks.

```python
sns.set(rc={'axes.facecolor': 'gray', 'figure.facecolor':'gray'})
spotifyPalette = sns.color_palette(["#1DB954", "#656966", "#d0d6d1"])
sns.set_palette(spotifyPalette)
plt.pie(albumCounts, labels=None, autopct='%1.3f%%', textprops={'color': '#052109'})
plt.title('Album Types', color='#010d03', fontsize=26)
custom_labels = ['Album', 'Single', 'Composition']
plt.legend(custom_labels, bbox_to_anchor=(.72, 0.09))
plt.show()
```

9) Analyze songs by artist and views-youtube, stream-spotify.

```python
analyzeSongByArtistViewsStream = df.groupby('Artist')[['Views', 'Stream']].sum()
top20Views = analyzeSongByArtistViewsStream.sort_values('Views', ascending=False).head(20).reset_index()
top20Streams = analyzeSongByArtistViewsStream.sort_values('Stream', ascending=False).head(20).reset_index()
topArtists = top20Views.merge(top20Streams, left_index=True, right_index=True, suffixes=('-YouTube', '-Spotify'))
topArtists.index = topArtists.index + 1
topArtists.drop(columns=['Stream-YouTube', 'Views-Spotify'])
```

10) Create bar charts for Top 10 Artists on Youtube and Top 10 Artists on Spotify in 1 row of display.

```python
top10Artists = topArtists.head(10)
fig, ax = plt.subplots(1, 2, figsize=(15, 6))
top10Artists.plot(kind='bar', x='Artist-YouTube', y='Views-YouTube', ax=ax[0], legend=False)
ax[0].set_title('Top 20 artists by Youtube views')
ax[0].set_xlabel('Artist')
ax[0].set_ylabel('Views')
ax[0].yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.0f}'.format(x)))
top10Artists.plot(kind='bar', x='Artist-Spotify', y='Stream-Spotify', ax=ax[1], legend=False)
ax[1].set_title('Top 20 artists by Spotify streams')
ax[1].set_xlabel('Artist')
ax[1].set_ylabel('Streams')
ax[1].yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.0f}'.format(x)))
plt.tight_layout()
plt.show()
```

11) Make some remarks about the Top 10 Artists on Youtube and Top 10 Artists on Spotify.

12) Analyze Top 10 songs on the basis of views, comments, and likes: with titles Top 10 Songs with Most Views, Top 10 Songs with Most Comments, and Top 10 Songs with Most Likes.

```python
songsByLikesViewsComments = df.groupby('Track')[['Likes', 'Views', 'Comments']].sum()
songsByLikesViewsComments = songsByLikesViewsComments.apply(lambda x: x/1000000)
top10ByLikes = songsByLikesViewsComments.sort_values('Likes', ascending=False).head(10).reset_index()
top10ByViews = songsByLikesViewsComments.sort_values('Views', ascending=False).head(10).reset_index()
top10ByComments = songsByLikesViewsComments.sort_values('Comments', ascending=False).head(10).reset_index()
fig, ax = plt.subplots(3, 1, figsize=(10, 15))

top10ByLikes.plot(kind='barh', x='Track', y='Likes', ax=ax[0], legend=False)
ax[0].set_title('Top 10 Songs with Most Likes')
ax[0].set_xlabel('Likes (in Millions)')
ax[0].set_ylabel('Title')
ax[0].xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.2f}'.format(x)))

top10ByViews.plot(kind='barh', x='Track', y='Views', ax=ax[1], legend=False)
ax[1].set_title('Top 10 Songs with Most Views')
ax[1].set_xlabel('Views (in Billions)')
ax[1].set_ylabel('Title')
ax[1].xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.2f}'.format(x/1000)))


top10ByComments.plot(kind='barh', x='Track', y='Comments', ax=ax[2], legend=False)
ax[2].set_title('Top 10 Songs with Most Comments')
ax[2].set_xlabel('Comments (in Millions)')
ax[2].set_ylabel('Title')
ax[2].xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.2f}'.format(x)))

plt.tight_layout()
plt.show()
```

12) Make some remarks about the Top 10 Songs with Most Views, Top 10 Songs with Most Comments, and Top 10 Songs with Most Likes.

13) Analyzing the songs on the basis of Energy and Danceability and display a table of it.

```python
songsByEnergyDanceability = df.groupby('Track')[['Energy', 'Danceability']].sum()
top20ByEnergy = songsByEnergyDanceability.sort_values('Energy', ascending=False).head(20).reset_index()
top20ByDanceability = songsByEnergyDanceability.sort_values('Danceability', ascending=False).head(20).reset_index()
topSongs = top20ByEnergy.merge(top20ByDanceability, left_index=True, right_index=True, suffixes=('-Energy', '-Danceability'))
topSongs.index = topSongs.index + 1
topSongs = topSongs.drop(columns=['Danceability-Energy', 'Energy-Danceability'])
topSongs = topSongs.rename(columns={'Energy-Energy': 'Energy', 'Danceability-Danceability': 'Danceability'})

topSongs
```

15) Display a bar chart of the Top 10 Songs by Stream with their Energy, Danceability, and Acousticness and make some remarks about it.

```python
songsByStreamEnergyDanceabilityAccousticness = df.groupby('Track')[['Stream', 'Energy', 'Danceability', 'Acousticness']].sum()
top10SongsByStream = songsByStreamEnergyDanceabilityAccousticness.sort_values('Stream', ascending=False).head(10).reset_index()
top10SongsByStreamNormalized = top10SongsByStream.copy()
columnsToNormalize = ['Stream', 'Energy', 'Danceability', 'Acousticness']

for column in columnsToNormalize:
    top10SongsByStreamNormalized[column] = (top10SongsByStream[column] - top10SongsByStream[column].min()) / (top10SongsByStream[column].max() - top10SongsByStream[column].min())

x = top10SongsByStream['Track']
y1 = top10SongsByStreamNormalized['Stream']
y2 = top10SongsByStreamNormalized['Energy']
y3 = top10SongsByStreamNormalized['Danceability']
y4 = top10SongsByStreamNormalized['Acousticness']
width = 0.2
pos1 = np.arange(len(x))
pos2 = [x + width for x in pos1]
pos3 = [x + width*2 for x in pos1]
pos4 = [x + width*3 for x in pos1]
plt.bar(pos1, y1, width=width, label='Stream')
plt.bar(pos2, y2, width=width, label='Energy')
plt.bar(pos3, y3, width=width, label='Danceability')
plt.bar(pos4, y4, width=width, label='Acousticness', color='#416e4c')
plt.title('Top 10 Songs with their Energy, Danceability, and Accousticness (normalized)')
plt.xlabel('Track')
plt.ylabel('Normalized Value')
plt.legend(bbox_to_anchor=(1, 1))
plt.xticks(pos2, x, rotation=90)
plt.show()
```

Note: I have normalized the columns (ranging from 0 to 1) to avoid the number of streams overshadowing the other y values. If you wanna look at the actual table with their actual values:

16) Determine the top 10 artists on the Spotify platform whose songs have more likes and display a bar chart with a title Number of Likes per Artist and make some remarks.

```python
artistsByStreamLikesViews = df.groupby('Artist')[['Likes', 'Views']].sum()
top10ArtistByLikes = artistsByStreamLikesViews.sort_values('Likes', ascending=False).head(10).reset_index()
plt.figure(figsize=(10, 5))
plt.bar(top10ArtistByLikes['Artist'], top10ArtistByLikes['Likes'])
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.0f}'.format(x)))
plt.title('Number of Likes per Artist')
plt.xlabel('Artist')
plt.ylabel('Likes')
plt.xticks(rotation=90)
plt.show()
```

19) Exploring compilation releases by artists and identify top performers based on views and likes and display a table for each artist with their Views and Likes as the other columns. Make some remark about the table.

```python
dfByCompilationsOnly = df[df['Album_type'] == 'compilation']
artistByArtistCompilationsOnly = dfByCompilationsOnly.groupby('Artist')[['Likes', 'Views']].sum()
top20ArtistByLikesCompilations = artistByArtistCompilationsOnly.sort_values('Likes', ascending=False).head(20).reset_index()
top20ArtistByViewsCompilations = artistByArtistCompilationsOnly.sort_values('Views', ascending=False).head(20).reset_index()
topArtistsCompilations = top20ArtistByLikesCompilations.merge(top20ArtistByViewsCompilations, left_index=True, right_index=True, suffixes=('-Likes', '-Views'))
topArtistsCompilations.index = topArtistsCompilations.index + 1
topArtistsCompilations = topArtistsCompilations.drop(columns=['Views-Likes', 'Likes-Views'])
topArtistsCompilations = topArtistsCompilations.rename(columns={'Likes-Likes': 'Likes', 'Views-Views': 'Views'})

topArtistsCompilations
```

Once again, the ranking of the high-performing artists has changed, providing further evidence for the point made in the previous item.

20) Identify the top 10 channels based on views and display a bar chart of it and make some remarks.

```python
channelByViews = df.groupby('Channel')[['Views']].sum()
top10ChannelsByViews = channelByViews.sort_values('Views', ascending=False).head(10).reset_index()
plt.figure(figsize=(10, 5))
plt.bar(top10ChannelsByViews['Channel'], top10ChannelsByViews['Views'])
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.0f}'.format(x)))
plt.title('Top Channels Ranking Based on views')
plt.xlabel('Artist')
plt.ylabel('Views')
plt.xticks(rotation=90)
plt.show()
```

The sheer number of views that music channels can amass, often reaching into the billions, highlights the transformative impact that YouTube has had on how music is consumed. This phenomenon underscores not only the evolution of technology but also the transformation of the music industry as a whole.
